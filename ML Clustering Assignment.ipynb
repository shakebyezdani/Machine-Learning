In this assignment students have to compress racoon grey scale image into
5 clusters. In the end, visualize both raw and compressed image and look
for quality difference.

```python
import pandas as pd
import numpy as np
import scipy
from sklearn.cluster import KMeans
from scipy import misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Visualize gray scale image

f = scipy.misc.face(gray = True) 
"""
Get a 1024 x 768, color image of a raccoon face(when gray = False).
f  gives array of pixel values.

"""
plt.figure(figsize = (10,6))
plt.imshow(f, cmap = 'gray',vmin = 0, vmax= 255)  # Display data as an image, i.e., on a 2D regular raster.
plt.show()
```

```python
# Compressing gray scale image into 5 clusters
rows = f.shape[0]
cols = f.shape[1]
```

```python
image = f.reshape(rows*cols,1)     # Unstacking image pixels values to feed into KMeans algorithm
kmeans = KMeans(n_clusters = 5)
kmeans.fit(image)
```

```python
clusters = kmeans.cluster_centers_
labels = kmeans.labels_
labels = labels.reshape(rows,cols)             # stacking clustered labeled data into image data
plt.imsave('compressed_racoon.png',labels)
```

```python
#Visualize the compressed image
image= plt.imread('compressed_racoon.png')     # convert image into array of pixel values
plt.figure(figsize = (10,6))
plt.imshow(image)
plt.show()
```
